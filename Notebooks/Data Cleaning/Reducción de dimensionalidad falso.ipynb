## Quitamos valores NaN

```python
dfcopy.isna().sum()
```

```python
# Encontrar nombres de columnas con NaN
nan_columns_list = [col for col in df.columns if df[col].isna().any()]

# Mostrar la lista de columnas
print(nan_columns_list)
```

```python
# Contar el número de NaN en cada columna
nan_count = df.isnull().sum()

# Filtrar columnas donde el número de NaN es mayor que 10
columns_with_more_than_10_nans = nan_count[nan_count > 40000]

len(columns_with_more_than_10_nans)
```

```python
# Ordenar las columnas por la cantidad de valores NaN
sorted_columns_by_nan_count = nan_count.sort_values(ascending=False)

sorted_columns_by_nan_count
```

```python
correlation_matrix = dfcopy.corr().abs()
```

```python
upper_triangle = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))
```

```python
to_drop = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.9)]
```

```python
df_dropped = dfcopy.drop(dfcopy[to_drop], axis=1)
```
